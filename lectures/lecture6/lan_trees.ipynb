```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # để trực quan hoá đồ thị hình cây
# maplotlib inline: có thể dùng code này để các lệnh trực quan hoá phía sau không phải dùng "plt.show"
import seaborn as sns
```

```python
iris_df = sns.load_dataset('iris')
iris_df.head(5)
```

```python
# bảng dữ liệu này là cân bằng, vì số lượng các nhãn bằng nhau:
iris_df.species.value_counts()
```

```python
sns.countplot(data=iris_df, x='species')
plt.show()
```

```python
# trực quan hoá mối quan hệ giữa các biến, có thể sử dụng corellation metrix để xem mối quan hệ : 
sns.scatterplot(data=iris_df, 
                x='sepal_length',
                y='sepal_width',
                hue = 'species')
plt.show()
```

```python
# numerirc dữ liệu dạng catagorical: 
# thay vì sử dụng iris_df.species.map({'sentosa':1,'versicolor':2,'virginica':2}) thì có thể sử dụng thư viện sau trong sklearn: 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # gọi mô hình
le.fit(iris_df.species.unique()) # fit dữ liệu.
iris_df.species = le.transform(iris_df.species) # phải gắn hàm với một tên biến, thì mới chạy được lệnh
iris_df.species.unique()
# MAP THÌ PHÙ HỢP VỚI ÍT LABELS, CÒN `LabelEncoder` thì không phù hợp nếu muốn gán giá trị theo ý chủ đích cá nhân. 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, classification_report, plot_confusion_matrix
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(iris_df[['sepal_length','sepal_width']], iris_df['species'], test_size=0.2, random_state=42)
```

```python
tree = DecisionTreeClassifier(max_depth=5, 
                              max_features=2, # max_depth quy định cây có bao nhiêu tầng, max_future:bao nhiêu feature được chọn
                             criterion='entropy') # criterion để chọn phương pháp chia cây
tree.fit(xtrain, ytrain)
```

```python
y_pred = tree.predict(xtest)
print(classification_report(ytest, y_pred))
```

```python
confusion_matrix(ytest, y_pred)
# kết quả số là một array vuông, trong đó số chiều là số nhãn
```

```python
# để TRỰC QUAN HOÁ confusion_matrix: 
plot_confusion_matrix(tree, xtest, ytest)
plt.show()
```

```python
def visualize_classifier(model, X, y, ax=None, cmap='rainbow'):
    X, y = X.values, y.values
    ax = ax or plt.gca()
    
    # Plot the training points
    ax.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=cmap,
               clim=(y.min(), y.max()), 
               zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # fit the estimator
    xx, yy = np.meshgrid(np.linspace(*xlim, num=200),
                         np.linspace(*ylim, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

    # Create a color plot with the results
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap=cmap, zorder=1)

    ax.set(xlim=xlim, ylim=ylim)
    plt.show()
```

```python
visualize_classifier(tree, iris_df[['sepal_length','sepal_width']], iris_df['species'])
```

```python
#plot tree: lưu ý, tham số 'feature name' là feature train. 
plt.figure(figsize=(25,10)) # để tạo kích thước cho cây. 
_ = plot_tree(tree, feature_names=xtrain.columns, max_depth=5, filled=True) #filled = True để làm cho đồ thì có màu
plt.show()
```

## Thuật toán SVM = supported vector machine
- là nền tảng của ML
- phục vụ bải toàn 2 nhãn, (binary)

```python
iris_df.species.value_counts()
```

```python
# sử dụng hàm replace
# biến đổi tất cả giá trị lớn hơn 0:  sử dụng để ép kiểu và gán dữ liệu 
iris_df.species.gt(0).astype(int).value_counts() #greater than 0
#gt:>
#ge:>=
#eq: ==
# ne:!=
#lt: <
#ln:<=
```

```python
iris_df['target']=iris_df.species.gt(0).astype(int)
iris_df.tail()
```

```python
sns.scatterplot(data=iris_df,
               x='petal_length',
               y='petal_width',
               hue='target')
plt.show()
```

```python
X,y = iris_df[['petal_length','petal_width']],iris_df['target']
X_train, x_test, y_train, y_test=train_test_split(X, y, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
clf=SVC(kernel='linear')
clf.fit(X_train, y_train)
```

```python
pred=clf.predict(x_test)
```

```python
print(classification_report(y_test,pred))
```

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=100, linewidth=1, facecolor='none', edgecolor='black');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    plt.show()
```

```python
ax = sns.scatterplot(data=iris_df, x='petal_length', y='petal_width', hue='target', s=100, alpha=0.7, palette='tab10')
plot_svc_decision_function(clf, ax=ax, plot_support=True)
plt.show()
```
